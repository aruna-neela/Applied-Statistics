## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Drop rows with missing values
df.dropna(inplace=True)

# Convert categorical variables to appropriate data types
categorical_cols = ['hotel', 'meal', 'country', 'market_segment', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type', 'reservation_status']
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Convert date columns to datetime data type
date_cols = ['reservation_status_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col])

# Drop irrelevant columns
irrelevant_cols = ['agent', 'company']
df.drop(irrelevant_cols, axis=1, inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

# Display the first few rows of the cleaned dataset
print("Cleaned Dataset:")
print(df.head())
````

Manipulations:

4. Removed irrelevant columns and duplicates.

#### Chart - 4

````python
# Chart - 4 visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Extract month from reservation_status_date
df['reservation_month'] = df['reservation_status_date'].dt.month

# Count cancellations by month
cancellation_counts = df[df['is_canceled'] == 1]['reservation_month'].value_counts().sort_index()

# Chart-4: Trend of Booking Cancellations Over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x=cancellation_counts.index, y=cancellation_counts.values, marker='o')
plt.title('Trend of Booking Cancellations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Cancellations')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a line plot to visualize the trend of booking cancellations over time because it effectively displays the sequential nature of the data (months) and allows for easy identification of trends or patterns in cancellation behavior over different time periods.Answer Here.

##### 2. What is/are the insight(s) found from the chart?

From the chart, we can observe the following insights:

1. Booking cancellations tend to fluctuate throughout the year, with higher peaks observed in certain months.
2. There may be seasonal patterns or trends in cancellation behavior, such as higher cancellations during peak travel seasons or specific events.

#### Chart - 5 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a correlation heatmap because it effectively visualizes the pairwise correlations between numerical variables in the dataset. This type of chart allows for a quick and comprehensive understanding of the relationships between variables, highlighting potential patterns or dependencies that may exist.Answer Here.

##### 2. What is/are the insight(s) found from the chart?

From the correlation heatmap, we can observe the following insights:

1. There is a strong positive correlation between variables such as lead time and booking cancellations, indicating that bookings made further in advance are more likely to be canceled.
2. There is a negative correlation between variables such as lead time and stays in weekend nights, suggesting that longer lead times may be associated with fewer weekend stays.
3. Variables such as average daily rate (ADR) and total of special requests show weak correlations with other numerical variables, indicating that they may be influenced by factors not captured in the dataset.

These insights provide valuable information about the relationships between numerical variables in the dataset, which can inform further analysis and strategic decision-making.
